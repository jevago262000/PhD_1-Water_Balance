```python
import pandas as pd
import numpy as np
```

```python
main_folder = r"Z:\PhD_Datasets&Analysis\Info_Inputs"
css_folder = main_folder + "\\Streamflow_Stations\Climate_Sensitive_Stations-GRDC"

filtered_css = pd.read_csv(css_folder + "//CSS_FINAL_SELECTION-MERGE_WITH_ALL.csv") # Read file with all CSS information
filtered_css
```

```python
monthly_folder = css_folder + "\\2025-02-13_17-18_Monthly"
obs_df = pd.read_csv(monthly_folder + "//_DataFrames//Joined_Monthly_Sts_DFs_mm.csv", index_col="YYYY-MM") # Observed data in millimeters
obs_df.index = pd.to_datetime(obs_df.index, format="%Y-%m-%d").strftime("%Y-%m")  # Convert index to the right format
obs_df
```

```python
stats_df_final = pd.read_csv(css_folder + "//runoff_monthly_stats_based-on_monthly_k.csv", index_col="Station") # Statistics DataFrame based on monthly flow series determined using monthly recession contants (k)
stats_df_final
```

```python
wb_output_folder = r"Z:\PhD_Datasets&Analysis\Outputs\T&M_WBM"
bflow_df = pd.read_csv(wb_output_folder + "//bflow2//bflow2_zonal_statistics_1958-2023.csv", index_col="YYYY-MM") # Baseflow DataFrame
bflow_df
```

```python
bflow_ant_df = pd.read_csv(wb_output_folder + "//bflow2//bflow2_ant_zonal_statistics_1958-2023.csv", index_col="YYYY-MM") # Baseflow DataFrame
bflow_ant_df
```

```python
perc_df = pd.read_csv(wb_output_folder + "//perc//perc_zonal_statistics_1958-2023.csv", index_col="YYYY-MM") # Baseflow DataFrame
perc_df
```

```python
monthly_k_recessions = pd.read_csv(monthly_folder + "//Baseflow//monthly_k_recessions_df.csv", index_col="station") # Monthly recession constants (k) DataFrame
monthly_k_recessions.index.name = "Station" # Set index name to "Station"
monthly_k_recessions
```

```python
non_containing_ws = pd.read_csv(main_folder + "\Streamflow_Sts_Drainage_Areas\GRDC_Watersheds\\non_containing_grdcno_int.csv", index_col="grdcno_int") # Watersheds not containing other GRDC watersheds
non_containing_ws
```

```python
def k_calculation(pbias, obs_str, perc, bflow_ant):
    """Calculates the monthly k based on PBIAS analysis."""
    pbias_fraction = -(pbias/100)
    return (np.sum(pbias_fraction * obs_str) - np.sum(perc)) / (np.sum(bflow_ant) - np.sum(perc))
```

```python
import warnings
warnings.filterwarnings('ignore')

sts_adjusted = []
for st in monthly_k_recessions.index:
    st_str = str(st)

    # Check if station exists in all required DataFrames
    if (
        st_str in obs_df.columns
        and st_str in perc_df.columns
        and st_str in bflow_ant_df.columns
        and st in stats_df_final.index
        and st in non_containing_ws.index
    ):
        # Get the relevant columns as Series
        obs_series = obs_df[st_str]
        perc_series = perc_df[st_str]
        bflow_ant_series = bflow_ant_df[st_str]

        # Create a DataFrame to align and drop rows with any NaNs
        aligned_df = pd.concat([obs_series, perc_series, bflow_ant_series], axis=1, keys=['obs', 'perc', 'bflow_ant']).dropna()

        obs_str = aligned_df['obs'].values
        perc = aligned_df['perc'].values
        bflow_ant = aligned_df['bflow_ant'].values
        pbias = stats_df_final.loc[st, "PBIAS"]

        k_value = k_calculation(pbias, obs_str, perc, bflow_ant)

        if not np.isnan(np.log(k_value)) and (k_value < 1):
            # If k_value is valid, update the DataFrame
            monthly_k_recessions.loc[st, "alpha_with_weights_adjusted"] = abs(np.log(k_value))
            monthly_k_recessions.loc[st, "k_recession_adjusted"] = k_value
            sts_adjusted.append(st)
        else:
            monthly_k_recessions.loc[st, "alpha_with_weights_adjusted"] = monthly_k_recessions.loc[st, "alpha_with_weights"]
            monthly_k_recessions.loc[st, "k_recession_adjusted"] = monthly_k_recessions.loc[st, "k_recession"]

    else:
        # If any required data is missing, set as the default value
        monthly_k_recessions.loc[st, "alpha_with_weights_adjusted"] = monthly_k_recessions.loc[st, "alpha_with_weights"]
        monthly_k_recessions.loc[st, "k_recession_adjusted"] = monthly_k_recessions.loc[st, "k_recession"]

print(f"Stations adjusted {len(sts_adjusted)}\n\t{sts_adjusted}")
monthly_k_recessions
```
