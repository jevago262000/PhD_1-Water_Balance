```python
import pandas as pd
```

```python
excel_folder = r"Z:\PhD_Datasets&Analysis\Info_Inputs\Streamflow_Stations\_DataFrames"
excel_file = "DFs.xlsx"
excel_path = excel_folder + "\\" + excel_file
df = pd.ExcelFile(excel_path)
sheet_names = df.sheet_names
print(sheet_names)
```

```python
terra_st_yr = 1958
terra_ed_yr = 2023
years = range(terra_st_yr, terra_ed_yr + 1)
months = range(1,12 + 1)
yr_mth= [str(a)+ "-" + str(b) for a in years for b in months]
df_sts = pd.DataFrame({"yr-mth": yr_mth}).set_index("yr-mth")
df_sts
```

```python
for st_name in sheet_names:
    # Read Excel file into a Pandas DataFrame
    df = pd.read_excel(excel_path, sheet_name=st_name).set_index("yr-mth")
    df_sts = pd.concat([df_sts, df], axis=1)

df_sts.to_csv(excel_folder + "\Joined_Sts_DFs.csv")
df_sts
```
