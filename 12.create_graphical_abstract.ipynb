```python
#!/usr/bin/env python3
"""
Horizontal Graphical Abstract with Manuscript Figures - Journal of Hydrology
Landscape: 13 cm × 5 cm (1328 × 531 pixels at 300 DPI)
Incorporates Fig3, Fig4, and Fig6 from the manuscript
"""

import os
import docx
from pptx import Presentation
from pptx.util import Cm, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN
from pptx.enum.shapes import MSO_SHAPE

# Read manuscript
print("Reading manuscript...")
doc = docx.Document(r'D:\OneDrive - CGIAR\Documents\PhD_JLU Giessen\Papers\Paper1\Submission\Journal_of_Hydrology\Open-Source Spatially Distributed Global Monthly Water Balance Model.docx')

print("\nCreating HORIZONTAL graphical abstract with manuscript figures...")

# Figure paths
fig3_path = r'D:\OneDrive - CGIAR\Documents\PhD_JLU Giessen\Papers\Paper1\Figures\Fig3.png'
fig4_path = r'D:\OneDrive - CGIAR\Documents\PhD_JLU Giessen\Papers\Paper1\Figures\Fig4.png'
fig6_path = r'D:\OneDrive - CGIAR\Documents\PhD_JLU Giessen\Papers\Paper1\Figures\Fig6.png'

# Create presentation - LANDSCAPE (13 cm × 5 cm)
prs = Presentation()
prs.slide_width = Cm(13)
prs.slide_height = Cm(5)
slide = prs.slides.add_slide(prs.slide_layouts[6])

# Color palette
BLUE = RGBColor(0x1E, 0x88, 0xE5)
TEAL = RGBColor(0x00, 0xAC, 0xC1)
GREEN = RGBColor(0x43, 0xA0, 0x47)
ORANGE = RGBColor(0xFF, 0x98, 0x00)
WHITE = RGBColor(0xFF, 0xFF, 0xFF)
DARK = RGBColor(0x2C, 0x3E, 0x50)
LIGHT = RGBColor(0xF5, 0xF5, 0xF5)

def box(l,t,w,h,fc,lc=None):
    sh=slide.shapes.add_shape(MSO_SHAPE.ROUNDED_RECTANGLE,l,t,w,h)
    sh.fill.solid();sh.fill.fore_color.rgb=fc
    if lc:sh.line.color.rgb=lc;sh.line.width=Pt(1.5)
    else:sh.line.fill.background()
    return sh

def txt(l,t,w,h,text,size=8,bold=False,color=DARK,align=PP_ALIGN.CENTER):
    tb=slide.shapes.add_textbox(l,t,w,h);tf=tb.text_frame;tf.word_wrap=True
    tf.margin_top=Pt(2);tf.margin_bottom=Pt(2);tf.margin_left=Pt(3);tf.margin_right=Pt(3)
    p=tf.paragraphs[0];p.text=text;p.font.size=Pt(size);p.font.bold=bold
    p.font.color.rgb=color;p.font.name='Arial';p.alignment=align;p.line_spacing=1.0
    return tb

def add_image(img_path, l, t, w, h):
    """Add image to slide"""
    return slide.shapes.add_picture(img_path, l, t, width=w, height=h)

def arrow_right(x, y):
    """Add right arrow"""
    sh = slide.shapes.add_shape(MSO_SHAPE.RIGHT_ARROW, Cm(x), Cm(y), Cm(0.4), Cm(0.2))
    sh.fill.solid(); sh.fill.fore_color.rgb = DARK
    sh.line.fill.background()
    return sh

# White background
box(Cm(0), Cm(0), Cm(13), Cm(5), WHITE)

# === TITLE BAR ===
box(Cm(0), Cm(0), Cm(13), Cm(0.45), BLUE)
txt(Cm(0.1), Cm(0.03), Cm(12.8), Cm(0.38),
    "Open-Source Spatially Distributed Global Monthly Water Balance Model with Baseflow Recession Module",
    9, True, WHITE)

# === SECTION 1: INPUTS with Fig3 (0.6-3.2 cm width) ===
box(Cm(0.1), Cm(0.55), Cm(3.0), Cm(0.3), TEAL)
txt(Cm(0.1), Cm(0.57), Cm(3.0), Cm(0.26), "GLOBAL OPEN DATASETS", 7.5, True, WHITE)

# Add Fig3 (soil parameters)
def add_image(img_path, l, t, width=None, height=None):
    """Add image to slide"""
    return slide.shapes.add_picture(img_path, l, t, width=width, height=height)

txt(Cm(0.2), Cm(2.75), Cm(2.8), Cm(0.35),
    "TerraClimate • SoilGrids 2.0 • CSIRO\n588 CSS Pristine Stations (1958-2023)",
    6, False, DARK)

# Arrow
arrow_right(3.15, 1.9)

# === SECTION 2: MODEL (3.6-5.8 cm width) ===
box(Cm(3.6), Cm(0.55), Cm(2.1), Cm(0.3), BLUE)
txt(Cm(3.6), Cm(0.57), Cm(2.1), Cm(0.26), "WATER BALANCE MODEL", 7.5, True, WHITE)

box(Cm(3.65), Cm(0.9), Cm(2.0), Cm(2.2), LIGHT, BLUE)

txt(Cm(3.7), Cm(0.95), Cm(1.9), Cm(0.25), "P = ET + Q + ΔS", 9, True, BLUE)
txt(Cm(3.7), Cm(1.25), Cm(1.9), Cm(0.2), "Thornthwaite-Mather Framework", 6, False, DARK)

# Two-component boxes
box(Cm(3.75), Cm(1.5), Cm(0.85), Cm(0.35), RGBColor(0xE3,0xF2,0xFD), BLUE)
txt(Cm(3.75), Cm(1.52), Cm(0.85), Cm(0.3), "Surface\nRunoff", 5.5, False, DARK)

box(Cm(4.7), Cm(1.5), Cm(0.85), Cm(0.35), RGBColor(0xB2,0xDF,0xDB), TEAL)
txt(Cm(4.7), Cm(1.52), Cm(0.85), Cm(0.3), "⭐ Baseflow\nRecession", 5.5, True, DARK)

txt(Cm(3.7), Cm(1.9), Cm(1.9), Cm(0.25), "Monthly MRC Fitting", 6.5, False, DARK)
txt(Cm(3.7), Cm(2.15), Cm(1.9), Cm(0.2), "R² = 0.98", 7, True, GREEN)

txt(Cm(3.7), Cm(2.4), Cm(1.9), Cm(0.6),
    "Python + ArcPy\nGoogle Earth Engine\nOpen Source Code",
    5.5, False, DARK)

# Arrow
arrow_right(5.75, 1.9)

# === SECTION 3: VALIDATION with Fig4 (6.2-8.8 cm width) ===
box(Cm(6.2), Cm(0.55), Cm(2.5), Cm(0.3), GREEN)
txt(Cm(6.2), Cm(0.57), Cm(2.5), Cm(0.26), "GLOBAL VALIDATION", 7.5, True, WHITE)

# Add Fig4 (validation basins by continent)
add_image(fig4_path, Cm(6.25), Cm(0.9), width=Cm(2.4), height=Cm(1.8))

txt(Cm(6.3), Cm(2.75), Cm(2.3), Cm(0.35),
    "N.America: 357 • Europe: 162\nOceania: 46 • S.America: 23",
    6, False, DARK)

# Arrow
arrow_right(8.75, 1.9)

# === SECTION 4: RESULTS with Fig6 (9.2-12.9 cm width) ===
box(Cm(9.2), Cm(0.55), Cm(3.6), Cm(0.3), ORANGE)
txt(Cm(9.2), Cm(0.57), Cm(3.6), Cm(0.26), "MODEL OUTPUTS & KEY RESULTS", 7.5, True, WHITE)

# Add Fig6 (TerraClimate Runoff, Baseflow, Total Runoff)
add_image(fig6_path, Cm(9.25), Cm(0.9), width=Cm(3.5), height=Cm(1.4))

# Performance metrics below maps
box(Cm(9.3), Cm(2.35), Cm(1.6), Cm(0.4), RGBColor(0xFF,0xF3,0xE0), ORANGE)
txt(Cm(9.35), Cm(2.4), Cm(1.5), Cm(0.3),
    "KGE: -0.07 → +0.22\nBias: -26.5% → +1.9%", 6, True, GREEN)

box(Cm(9.3), Cm(2.8), Cm(1.6), Cm(0.35), RGBColor(0xE0,0xF2,0xF1), TEAL)
txt(Cm(9.35), Cm(2.85), Cm(1.5), Cm(0.25),
    "Subsurface: 5-42%\nof Continental Runoff", 6, True, TEAL)

box(Cm(11.0), Cm(2.35), Cm(1.75), Cm(0.8), LIGHT, GREEN)
txt(Cm(11.05), Cm(2.4), Cm(1.65), Cm(0.7),
    "Continental Runoff (mm/yr):\nS.Am: 674 • Europe: 414\nAsia: 298 • N.Am: 292\nAfrica: 198 • Oceania: 128",
    5.5, False, DARK)

# === FOOTER ===
box(Cm(0.1), Cm(4.55), Cm(12.8), Cm(0.4), DARK)
txt(Cm(0.2), Cm(4.58), Cm(4.0), Cm(0.33),
    "Temporal: 1958-2023 | Validation: 588 Pristine Basins", 6, False, WHITE, PP_ALIGN.LEFT)
txt(Cm(4.5), Cm(4.58), Cm(4.0), Cm(0.33),
    "100% Open Data • Open Code • Reproducible", 6.5, True, WHITE)
txt(Cm(8.8), Cm(4.58), Cm(3.8), Cm(0.33),
    "Python • GEE • Global Scale", 6, False, WHITE, PP_ALIGN.RIGHT)

# Save
output_path = r'D:\OneDrive - CGIAR\Documents\PhD_JLU Giessen\Papers\Paper1\Submission\Journal_of_Hydrology\Graphical_Abstract.pptx'
prs.save(output_path)

print("\n✓ SUCCESS: Created Graphical_Abstract.pptx")
print("  Format: HORIZONTAL 13×5 cm (1328×531 px)")
print("  Includes manuscript figures:")
print("    - Fig3: Global soil parameters (Input section)")
print("    - Fig4: Validation basins by continent (Validation section)")
print("    - Fig6: Model outputs - Runoff & Baseflow maps (Results section)")
print("  Layout: Input → Model → Validation → Results")
print("  Color-coded workflow with integrated maps")
print(f"\n  Saved to: {output_path}")
print("\nReady for Journal of Hydrology submission!")
```
