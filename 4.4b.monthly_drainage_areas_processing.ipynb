```python
import pandas as pd, glob, numpy as np
```

```python
css_folder = r"Z:\PhD_Datasets&Analysis\Info_Inputs\Streamflow_Stations\Climate_Sensitive_Stations-GRDC"
monthly_folder = css_folder + "\\2025-02-13_17-18_Monthly"

filtered_css = pd.read_csv(css_folder + "//CSS_FINAL_SELECTION-MERGE.csv") # Read file with filtered CSS stations
filtered_css
```

```python
stations = filtered_css["station_no"].tolist() # List of stations to be processed
sts_dict = {}

for station in stations:

    file_name = str(station) + "_Q_Month.txt"
    file_to_read = monthly_folder + "\\" + file_name
    print("Reading file: " + file_name)

    # Specify encoding explicitly
    with open(file_to_read, 'r', encoding='ISO-8859-1') as inFile:    
        info = inFile.readlines()[16] # line with Next downstream station
    
    ndss = info.split(":")[1].strip() # Split the line into a list, takes the second element and removes leading/trailing spaces

    if ndss == "-": # This avoids saving no-data values (-)
        ndss = np.nan

    sts_dict[station] = ndss # Add the station and its next downstream station to the dictionary

# Convert the dictionary to a DataFrame
df = pd.DataFrame.from_dict(sts_dict, orient='index', columns=['Next_Downstream_Station'])
df.index.name = "station_no"
df
```

```python
# Map the Next_Downstream_Station values from df to filtered_css based on station_no
filtered_css['Next_Downstream_Station'] = filtered_css['station_no'].map(df['Next_Downstream_Station'])
filtered_css
```

```python
# Split the CATCHMENT_SIZE field by " ", take the first part, and replace the comma with a dot
filtered_css['CATCHMENT_SIZE2'] = filtered_css['CATCHMENT_SIZE'].str.split(" ").str[0].str.replace(",", ".").astype(float)

# Sort the dataframe by CATCHMENT_SIZE2 in ascending order and reset the index
sorted_css = filtered_css.sort_values(by='CATCHMENT_SIZE2', ascending=True).reset_index()

# Assign Priority based on the order of CATCHMENT_SIZE2 (smaller size gets higher priority)
sorted_css['Priority'] = range(len(sorted_css), 0, -1)

# Create a new dataframe with the relevant columns
priority_df = sorted_css[['station_no', 'CATCHMENT_SIZE2', 'Priority']]
priority_df = priority_df.set_index('station_no')
priority_df
```

```python
# Map the Priority values from priority_df to filtered_css based on station_no
filtered_css['Priority'] = filtered_css['station_no'].map(priority_df['Priority'])
filtered_css
```

```python
# Save the final dataframe with all the needed data to a CSV file
filtered_css.to_csv(css_folder + "\\CSS_FINAL_SELECTION-MERGE_WITH_ALL.csv", index=False) 
filtered_css
```
